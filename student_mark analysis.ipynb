```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# re-arrange the index position after delete the null values
dfs.reset_index(inplace=True)
```

```python

dfs.drop(['index'],axis=1,inplace=True)
```

```python
dfs['ParentEduc'].unique()
```

```python
# change the parents education name for better analysis
dfs['ParentEduc']=dfs['ParentEduc'].map({"master's degree":'Master','some college':'col',"associate's degree":'As.dgr','high school':'school','some high school':'h.school',"bachelor's degree":'bachelor'})
```

**GENDER IMPACT--**
The analysis indicates a marginal performance difference between male and female students, with male students showing slightly higher average Math scores. However, the gap is not significantly large, suggesting relatively balanced academic performance across genders.

**ETHNIC GROUP INFLUENCE--**
Variations in Math scores across different ethnic groups suggest that socio-cultural or environmental factors may contribute to performance differences. Certain groups demonstrate comparatively higher averages, highlighting the importance of targeted academic support strategies.

**PARENTIAL EDUCATION IMPACT--**
A strong positive correlation is observed between parents’ education level and student Math performance. Students whose parents hold higher educational qualifications (such as master’s or bachelor’s degrees) tend to score higher. This indicates that academic environment and parental guidance play a crucial role in student success.

**TEST PREPARATION--**
Students who completed test preparation courses performed better than those who did not. This demonstrates the measurable impact of structured preparation programs on academic outcomes.

**PARENTAL MARATIAL STATUS--**
The data shows minimal variation in scores based on parental marital status, suggesting that this factor has comparatively lower direct impact on academic performance.

**SPORTS PRACTICE--**
Students who regularly participate in sports show slightly better performance compared to those who never practice. This indicates that balanced physical activity may positively influence cognitive performance and discipline.

**WEEKLY STUDY HOURS--**
A clear upward trend is observed between study hours and Math scores. Students studying more than 10 hours per week achieve the highest average scores. This confirms that consistent study time is a major determinant of academic achievement.

**OVERALL INSIGHTE**

From an institutional or educational management perspective, the analysis highlights that:

Academic performance is strongly influenced by parental education, study hours, and structured test preparation.

Encouraging regular study habits and balanced extracurricular activities can enhance performance.

Targeted support programs can be designed for lower-performing groups to ensure equitable educational outcomes.

This data-driven approach can assist educational institutions in making strategic decisions to improve student performance and overall academic quality.
