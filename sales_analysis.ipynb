```python
plt.figure(figsize=(15,10),facecolor='cyan')
plt.gcf().suptitle('SALES DATA ANALYSIS',fontweight='bold',color='red',fontsize=18)

plt.subplot(2,2,1)
plt.pie(dfs['Gender'].value_counts(),labels=dfs['Gender'].unique(),autopct='%0.2f%%',wedgeprops={'edgecolor':'black','linewidth':1.5}
        ,textprops={'color':'black'},explode=(0.09,0.0),radius=1.0)
plt.title('GENDER DISTRIBUTION',fontweight='bold',color='black')


plt.subplot(2,2,2)
c=sns.countplot(dfs['Gender'],color='blue',edgecolor='black',orientation='horizontal')
plt.title('GENDER COUNT',fontweight='bold',color='black')
plt.xlabel('count',fontweight='bold',color='black')
plt.ylabel('Gender',fontweight='bold',color='black')
plt.xticks(fontweight='bold',color='black')
plt.yticks(fontweight='bold',color='black')
plt.gca().set_facecolor('lightgreen')
for i in c.containers:
    c.bar_label(i,fontweight='bold',color='black',label_type='center')

plt.subplot(2,2,3)
k=sns.countplot(dfs['Marital'],color='blue',edgecolor='black',orientation='horizontal')
plt.title('MARITAL STATUS',fontweight='bold',color='black',fontsize=10)
plt.xlabel('count',fontweight='bold',color='black')
plt.ylabel('Marital',fontweight='bold',color='black')
plt.xticks(fontweight='bold',color='black')
plt.yticks(fontweight='bold',color='black')
plt.gca().set_facecolor('plum')
for i in k.containers:
    k.bar_label(i,fontweight='bold',color='black',label_type='center')


plt.subplot(2,2,4)
plt.pie(dfs['Marital'].value_counts(),labels=dfs['Marital'].unique(),autopct='%0.2f%%',wedgeprops={'edgecolor':'black','linewidth':1.5},explode=(0.09,0.0)
              ,textprops={'fontsize':10,'fontweight':'bold'})
plt.title('MARITAL DISTRIBUTION',fontweight='bold',color='black')






plt.savefig('sales.png',dpi=500)
plt.tight_layout()
plt.show()
```

# Conclusion >>>

This project presents a detailed analysis of sales data with a focus on gender and marital status distribution to understand customer demographics and purchasing behavior. The objective of this analysis was to identify dominant customer segments and provide data-driven insights that can support effective business strategies.

The gender distribution analysis indicates that female customers form the majority of the customer base, accounting for approximately 69.69% of total customers, while male customers represent 30.31%. This significant difference highlights the stronger engagement and participation of female customers in purchasing activities, making them a critical segment for targeted marketing and customer retention initiatives.

The gender count analysis further supports this observation, showing a substantially higher number of female customers compared to male customers. This suggests that business offerings, promotional strategies, and product designs are more aligned with the preferences and needs of female consumers.

Marital status analysis reveals that single customers constitute the larger portion of the customer base, contributing nearly 58%, while married customers account for about 42%. The higher participation of single customers may be attributed to greater spending flexibility and independent purchasing decisions.

Overall, the findings indicate that female and single customers are the most influential demographic segments in the analyzed sales data. These insights can help organizations enhance customer segmentation, personalized marketing, and demand forecasting. By leveraging demographic-based analysis, businesses can optimize sales strategies, improve customer satisfaction, and achieve long-term growth
